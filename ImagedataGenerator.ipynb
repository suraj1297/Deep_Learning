```
import keras
from keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

!unzip -uq "/content/drive/My Drive/bird.zip" -d "/content/bird"
```

```
# Loading Image
img = load_img("/content/bird/bird.jpg")

# Transforming image into array
data = img_to_array(img)
```

```
# expand dimension to one sample
samples = np.expand_dims(data,0)
samples.shape
```

```
# Create Image data Augmentation generator
datagen = ImageDataGenerator(width_shift_range=[-200,200])

# prepare iterator
it = datagen.flow(samples, batch_size=1)

plt.figure(figsize=(10,10))
count = 1
for i in it:
  
  if count == 10:
    break
  else:
    plt.subplot(3,3,count)
    i = i.astype('uint8')
    plt.imshow(i[0,0:])
  count +=1
```

```
#Create an Image Data augmentation generator
datagen = ImageDataGenerator(height_shift_range=[-100,100])

# Preparing iterator
it = datagen.flow(samples, batch_size=1)

# Plotting
plt.figure(figsize=(10,10))
count= 1
for i in it:
  if count ==10:
    break
  plt.subplot(3,3,count)
  i = i.astype("uint8")
  plt.imshow(i[0,0:])
  count += 1
```
